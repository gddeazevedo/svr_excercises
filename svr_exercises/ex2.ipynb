```python
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from helpers import synthetic_data


epsilon = 0.01
C = 10

X, y = synthetic_data.generate_linear_data(ang_coef=2.0, linear_coef=1.0, n_samples=100, with_noise=True, noise_std=1.0)

svr = SVR(kernel="linear", epsilon=epsilon, C=C)
svr.fit(X, y)
y_pred_svr = svr.predict(X)

linear_regression = LinearRegression()
linear_regression.fit(X, y)
y_pred_lr = linear_regression.predict(X)

print("----MSE----")
print(f"Support Vector Regression: {mean_squared_error(y, y_pred_svr)}")
print(f"Linear Regression        : {mean_squared_error(y, y_pred_lr)}")

print("\n----MAE----")
print(f"Support Vector Regression: {mean_absolute_error(y, y_pred_svr)}")
print(f"Linear Regression        : {mean_absolute_error(y, y_pred_lr)}")

print("\n----R2 Score----")
print(f"Support Vector Regression: {r2_score(y, y_pred_svr)}")
print(f"Linear Regression        : {r2_score(y, y_pred_lr)}")

print("\n----Coeficientes----")
print(f"Linear Regression: {linear_regression.coef_}")
print(f"SVR: {svr.coef_}")

print("\n----Intercept----")
print(f"Linear Regression: {linear_regression.intercept_}")
print(f"SVR: {svr.intercept_}")

'''Plot do SVR'''
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color="lightblue", label="Dados")
plt.plot(X, y_pred_svr, color="blue", label="SVR Reta")
plt.plot(X, y_pred_svr + epsilon, "k--", label="Margen +")
plt.plot(X, y_pred_svr - epsilon, "k--", label="Margem -")
plt.ylabel("y")
plt.xlabel("X")
plt.title("SVR")
plt.legend()
plt.grid(True)
plt.show()

'''Plot do LinerRegression'''
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color="lightblue", label="Dados")
plt.plot(X, y_pred_lr, color="blue", label="LinearRegression Reta")
plt.ylabel("y")
plt.xlabel("X")
plt.title("LinerRegression")
plt.legend()
plt.grid(True)
plt.show()
```

<h4>Em qual cenrio o SVR mostra vantagens claras sobre o regressor linear? Por qu?</h4>

> O SVR  melhor que a regresso linear tradicional nos casos em que os dados so no lineares. Isso ocorre pois o SVR pode utilizar funes kernel, que mapeiam os dados para um espao de dimenso maior, o que permite entender melhor a estrutura dos dados e modelar padres mais complexos.
